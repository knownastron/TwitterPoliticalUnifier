```python
import csv
```

```python
file = open('congress_handles.csv')
readCsv = csv.reader(file, delimiter=',')
handles = []
for line in readCsv:
    name = line[0].rstrip().encode('ascii', 'ignore').decode("utf-8")
    handles.append([name, line[2].lower()])

file1 = open('PartyAffiliations.csv')
readCsv1 = csv.reader(file1, delimiter=',')
party = {}
for line in readCsv1:
    name = line[0].encode('ascii', 'ignore').decode("utf-8").rstrip()
    party[name] = line[1].rstrip()

```

```python
count = 1
for handle in handles:
    if party[handle[0]] == "Democratic Party":

        print(handle[1])
        count += 1
```
